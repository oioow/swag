**Q5:** In reality, ChatGPT uses *multi-head self attention*.  One way to achieve this is to split the weight matrices ($W_Q$,$W_K$ and $W_V$, see Lecture 3) along the output embedding dimension (i.e. their second dimension) into <code>noHeads</code> sections. The softmax function is applied to each section separately instead of as a whole.  Specifically, if we define Attention as:

$$\mathrm{Attention}(Q,K,V) = \mathrm{softmax}\left(\frac{QK^T}{\sqrt{d}}\right)V$$

where $d$ is the word embedding dimension (<code>embedDim</code>), then the multi-head attention attention is just an array:

$$\begin{bmatrix}
\mathrm{Attention}(Q_1,K_1,V_1)~~~~~
\mathrm{Attention}(Q_2,K_2,V_2)~~~~~ 
\cdots~~~~~
\mathrm{Attention}(Q_N,K_N,V_N)
\end{bmatrix}$$

Which can alternatively be expanded to:

$$\begin{bmatrix}
\mathrm{softmax}\left(\frac{Q_1K_1^T}{\sqrt{d/N}}\right)V_1~~~~~\mathrm{softmax}\left(\frac{Q_2K_2^T}{\sqrt{d/N}}\right)V_2~~~\cdots
~~~\mathrm{softmax}\left(\frac{Q_NK_N^T}{\sqrt{d/N}}\right)V_N
\end{bmatrix}$$

Where each matrix $Q_n$, $K_n$ and $V_n$ is split into $N$ chunks along the output embedding dimension. Please refer to Lecture 3 for a detailed description of how to implement multihead attention.

## Part 4: Training and testing the system [15%]
In this final part you will train and test the model. Much of the code to do this has been provided for you so that you can focus on answering questions about performance and concepts. Adjustments can be made by changing your parameters defined in Part 3.

```python
# The loop that is called for training
def train_loop(dataloader, model, loss_fn, optimiser):
    size = len(dataloader.dataset)

    # Set the model to training mode - important for batch normalization and dropout layers
    # Unnecessary in this situation but added for best practice
    model.train()
    for batch, (X, y) in enumerate(dataloader):
        pred = model(X)
        loss = loss_fn(pred, y)

        # Backpropagation
        loss.backward()
        optimiser.step()
        optimiser.zero_grad()

# Loop run each epoch to test performance on validation data
def test_loop(dataloader, model, loss_fn):
    # Set the model to evaluation mode - important for batch normalization and dropout layers
    # Unnecessary in this situation but added for best practices
    model.eval()
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss, correct = 0, 0

    # Evaluating the model with torch.no_grad() ensures that no gradients are computed during test mode
    # also serves to reduce unnecessary gradient computations and memory usage for tensors with requires_grad=True
    with torch.no_grad():
        for X, y in dataloader:
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()

    test_loss /= num_batches
    correct /= size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")

# Resent parameters so when the model is run it always starts afresh
for layer in model.children():
   if hasattr(layer, 'reset_parameters'):
       layer.reset_parameters()

# Code that actually trains and tests model by calling functions above.
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train_loop(train_loader, model, loss_fn, optimiser)
    test_loop(val_loader, model, loss_fn)

print("Done!")
```
